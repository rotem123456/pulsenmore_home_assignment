STEP FOUR: Data Preprocessing

```python
#lets the new_devices file
new_devices = pd.read_csv("new_devices.csv")
new_devices
```

At first glance, the data seems to be identical to the original sample group (Y1->Y5 & X2->X5) and we have no null data, so data preproccessing wont be neccesary
The assigment here is to rank these new devices based on their probabilty of belonging to the sample group
My first idea was to used the ML model K nearest neighbours, but given that our sample data are all in the same group and we dont have any given labels, I will try a differant approach
I will calculate the eucladian distance of each new_device to the mean of the samplegroup, and then rank the probability of them being part of the sample group based on the distance
in Pseudo:

for value in new_devices:
    dst <- distance of value from mean of sample group
    append dst to arry of distances

After we sort the array in ascending the ID with the lowest value has its distance closest to the mean of the sample group, therefore has a higher probability of belonging to the sample group

In order to this calculation in code, we will use the numpy and pandas frameworks
 

```python
#first lets calculate the centroid (mean value) of our original data
centroid = data.mean() #the mean value we recieve is a vector (for each value Y1 -> Y5 and the X values (which are constant values))
dist_arr = []
for i in range(new_devices.shape[0]):
    row = new_devices.iloc[i, 1:]  #I use iloc which is pandas method for getting the value of a row,here I use it to get the value of of each row from column 1 until the end(colum 0 is the ID) 
    distance = np.sqrt(((row - centroid) ** 2).sum())  #here I calculate the distance using the classic euclidian distance formula (substraction occurs beetwen two vectors)
    dist_arr.append(distance) #appending the distance to the distance array
dist_arr = np.array(dist_arr) #here we turn the array into a numpys array 

updated_new_devices = new_devices
updated_new_devices['Distance'] = dist_arr
updated_new_devices
```

In the code above a new table was called "updated new devices" and we added a Distance column to the given table.
As we can see in the Distance Column:
Device ID 2: This device has the smallest distance to the centroid of the sample group, which suggests that it's the most likely to belong to the sample group.
Device ID 3: This device has the second smallest distance to the centroid.
Device ID 1: This device has the largest distance to the centroid, indicating that it's the least likely to belong to the sample group among the three new devices.
